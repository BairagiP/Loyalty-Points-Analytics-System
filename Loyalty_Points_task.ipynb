# Part A - Calculating loyalty points

On each day, there are 2 slots for each of which the loyalty points are to be calculated:
S1 from 12am to 12pm 
S2 from 12pm to 12am

### 1. Find Playerwise Loyalty points earned by Players in the following slots:-<br>
    a. 2nd October Slot S1<br>
    b. 16th October Slot S2<br>
    b. 18th October Slot S1<br>
    b. 26th October Slot S2

```python
# Filter data for specific dates and slots
specific_dates = [
    ('10-02', 'S1'),#month - day ,slot
    ('10-16', 'S2'),
    ('10-18', 'S1'),
    ('10-26', 'S2')
    
]
default_year = 2022

```

**The criteria to convert number of games played, deposits and withdrawal into points is given as below :**

```python
deposit_weight = 0.01
withdrawal_weight = 0.005
gameplay_weight = 0.2
deposit_vs_withdrawal_weight = 0.001

loyalty_points_results = []
```

```python
for date, slot in specific_dates:
    target_date = pd.to_datetime(f"{default_year}-{date}", format='%Y-%m-%d')#taking date in target_date as 2022-10-02
    # print(target_date)

    #Filtering data based on date and slot for user_gameplay_df and storing in gameplay_slot

    gameplay_slot = user_gameplay_df[
        (user_gameplay_df['Datetime'].dt.month == target_date.month) &
        (target_date.day == user_gameplay_df['Datetime'].dt.day ) &
        (user_gameplay_df['slot'] == slot)
    ]

    #Filtering data based on date and slot for deposit_df and storing in deposit_slot
    deposit_slot = deposit_df[
        (deposit_df['Datetime'].dt.month == target_date.month) &
        (deposit_df['Datetime'].dt.day == target_date.day) &
        (deposit_df['slot'] == slot)]
    
    #Filtering data based on date and slot for withdrawal_df and storing in withdrawal_slot
    withdrawal_slot = withdrawal_df[
        (withdrawal_df['Datetime'].dt.month == target_date.month) &
        (withdrawal_df['Datetime'].dt.day == target_date.day) &
        (withdrawal_df['slot'] == slot)]
    
    # Calculate loyalty points 

    for user_id in gameplay_slot['User ID'].unique():

        #getting games played,deposit_amount,withdrawal_amount for finding points based on them
        games_played = gameplay_slot[gameplay_slot['User ID'] == user_id]['Games Played'].sum()
        deposit_amount = deposit_slot[deposit_slot['User Id'] == user_id]['Amount'].sum()
        withdrawal_amount = withdrawal_slot[withdrawal_slot['User Id'] == user_id]['Amount'].sum()
        
        #getting num_deposits and num_withdrawals for finding bonus_points
        num_deposits = deposit_slot[deposit_slot['User Id'] == user_id].shape[0]
        num_withdrawals = withdrawal_slot[withdrawal_slot['User Id'] == user_id].shape[0]
        
        # Calculate points
        points_from_deposit = deposit_weight * deposit_amount
        points_from_withdrawal = withdrawal_weight * withdrawal_amount
        points_from_games = gameplay_weight * games_played
        bonus_points = deposit_vs_withdrawal_weight * max(num_deposits - num_withdrawals, 0)
        
        # Total points
        total_points = points_from_deposit + points_from_withdrawal + points_from_games + bonus_points
        
        # Store the results
        loyalty_points_results.append({
            'User Id': user_id,
            'Date': date,
            'Slot': slot,
            'Loyalty Points': total_points
        })

# Convert to DataFrame
loyalty_points_df = pd.DataFrame(loyalty_points_results)
```

### 2. Calculate overall loyalty points earned and rank players on the basis of loyalty points in the month of October. 
###     In case of tie, number of games played should be taken as the next criteria for ranking.

```python
#For loyalty points of al the users in october
loyalty_points_all_user = []

for user_id in user_gameplay_df['User ID'].unique():
    
    user_gameplay_october = user_gameplay_df[
        (user_gameplay_df['User ID'] == user_id) &
        (user_gameplay_df['Datetime'].dt.month == 10)
    ]
    
    user_deposit_october = deposit_df[
        (deposit_df['User Id'] == user_id) &
        (deposit_df['Datetime'].dt.month == 10)
    ]
    
    user_withdrawal_october = withdrawal_df[
        (withdrawal_df['User Id'] == user_id) &
        (withdrawal_df['Datetime'].dt.month == 10)
    ]
    
    games_played = user_gameplay_october['Games Played'].sum()
    deposit_amount = user_deposit_october['Amount'].sum()
    withdrawal_amount = user_withdrawal_october['Amount'].sum()
    
    num_deposits = user_deposit_october.shape[0]
    num_withdrawals = user_withdrawal_october.shape[0]
    
    # Calculate points
    points_from_deposit = deposit_weight * deposit_amount
    points_from_withdrawal = withdrawal_weight * withdrawal_amount
    points_from_games = gameplay_weight * games_played
    bonus_points = deposit_vs_withdrawal_weight * max(num_deposits - num_withdrawals, 0)
    
    total_points = points_from_deposit + points_from_withdrawal + points_from_games + bonus_points
    
    # Append the results to the list
    loyalty_points_all_user.append({
        'User Id': user_id,
        'Loyalty Points': total_points,
        'Games Played': games_played
    })


loyalty_points_df = pd.DataFrame(loyalty_points_all_user)

# Ranking by loyalty points and games played.
ranked_points = loyalty_points_df.sort_values(
    by=['Loyalty Points', 'Games Played'], 
    ascending=[False, False]
).reset_index(drop=True)

# Adding a ranking column
ranked_points['Rank'] = ranked_points.index + 1

# Displaying the top players
print(ranked_points)
```
